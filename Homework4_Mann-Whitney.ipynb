### Non parametric Mann-Whitney U test

##### Homework 4: Evaluate the Mann-Whitney U test for DIC and temperature. Create boxplots and violun plots for the analyses. 

```python
def import_csv (file):
    import pandas as pd
    return pd.read_csv (file)
```

```python
# Path to the file to be imported
path = "Course_data.csv"

# Import the file and convert to a dataframe
data_df = import_csv(path)
```

```python
DIC = data_df['dic_micromol_kg']
temp = data_df['temp_c']
```

```python
#Calling Mannwhitneyu
from scipy.stats import mannwhitneyu
```

```python
def mann_whitney_test (data1, data2, alpha= 0.05):
    result = mannwhitneyu(data1, data2, alternative= 'two-sided')

    print(f'U statistic = {result.statistic}')
    print(f'p-value = {result.pvalue}')

    if result.pvalue < alpha:
        print ('The difference is statistically significant')
    else:
        print ('The difference is not statistically significant')
```

#### DIC boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

colors = ['mediumslateblue']

sns.set(font_scale=2, style='ticks', palette= colors)
```

```python
sns.boxplot(data= data_df,
                 x= 'season', y= 'dic_micromol_kg')

plt.xlabel('Season', fontname= "Georgia", fontsize=12)
plt.ylabel ('Dissolved Inorganic Carbon (DIC)', fontname= "Georgia", fontsize= 12)
plt.title("Box plot for Dissolved Inorganic Carbon by season",
          fontname= "Georgia", fontsize = 14)
plt.show()
```

#### DIC violin plot

```python
sns.violinplot(data= data_df, x= 'season', y='dic_micromol_kg')

plt.xlabel('Season', fontname= "Georgia", fontsize=12)
plt.ylabel ('Dissolved Inorganic Carbon (DIC)', fontname= "Georgia", fontsize= 12)
plt.title("Violin plot for Dissolved Inorganic Carbon by season",
          fontname= "Georgia", fontsize = 14)
plt.show()
```

##### Perfoming Mann-Whitney U test for DIC

```python
DIC_dry = data_df.loc[data_df['season'] == 'Dry', 'dic_micromol_kg']
DIC_rainy = data_df.loc[data_df['season'] == 'Rainy', 'dic_micromol_kg']

mann_whitney_test(DIC_dry, DIC_rainy)
```

#### Perfoming Mann-Whitney U test for temperature

```python
temp_dry = data_df.loc[data_df['season'] == 'Dry', 'temp_c']
temp_rainy = data_df.loc[data_df['season'] == 'Rainy', 'temp_c']

mann_whitney_test(temp_dry, temp_rainy)
```
