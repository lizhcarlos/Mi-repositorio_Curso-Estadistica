## Homework 5: Histograms 

#### Histograms contructed from a geological database. The data correspond to Garnet Compactness values from a dune fiel of the SE Pacific.

```python
import pandas as pd

#Import Compactness data
C = pd.read_csv('Compacidad.csv')
```

```python
import matplotlib.pyplot as plt
fig, ax3 = plt.subplots()

colors = ['darksalmon', 'yellowgreen', 'powderblue', 'mediumslateblue','darkorchid', 'deeppink']
labels = ['La Colorada', 'Garrapatera', 'Concepción Bamba', 'Chipehua', 'Playa Azul', 'La Ventosa']


# Make a multiple-histogram of data-sets with different length.
ax3.hist(C, 6, histtype='bar', color = colors, label = labels)
ax3.set_title('Garnet compactness from different sample sizes', 
              fontname= "Georgia", fontsize = 14)
plt.xlabel ('Compactness', fontname= "Georgia", fontsize=12)
plt.ylabel ('Frecuency', fontname= "Georgia", fontsize= 12)
plt.legend(loc='upper left', fontsize= 10, prop= "Georgia", frameon= False)
plt.grid(True, alpha=0.2)

plt.show()
```
