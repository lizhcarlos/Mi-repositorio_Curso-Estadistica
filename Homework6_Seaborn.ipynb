### Homework 6: Seaborn

```python
import pandas as pd

#Import Compactness data
morphology = pd.read_csv('Garnet_morphology.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

colors = ['darksalmon', 'mediumslateblue','darkorchid', 'deeppink']

sns.set(font_scale=2, style='ticks', palette= colors)
```

#### *Boxplot with catplot in seaborn*

```python
ax = sns.catplot(data= morphology,
                 x= 'grain', y= 'compactness',
                 col="garnet",
                 hue= 'dune', kind= 'box',
                 height= 7, aspect=1)

plt.show()
```
